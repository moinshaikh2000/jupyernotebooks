```python
def check_age_group(age):
    if age <= 15:
        age_group = '0 - 15'
    elif (age > 15) & (age <=22):
        age_group = '16- 22'
    elif (age > 22) & (age <= 30):
        age_group = '23 - 30'
    else:
        age_group = '> 30'
    return age_group
```

```python
check_age_group(120)
```

```python
check_age_group(23)
```

```python
def login(user_name,password,user_db = user_db):
    login_status = False
    if user_name in user_db.keys():
        if user_db[user_name]  == password:
            login_status = True
        else:
            login_status = 1
    else:
        login_status = 2
    return login_status
```

```python
login('arshiyan','123')
```

```python
login('roshan','pass431')
```

```python
def add_nums(*args):
    _sum = 0
    for i in args:
        if isinstance(i,int) or isinstance(i,float):
            _sum += i
    return _sum
    
```

```python
add_nums(1,2,3,3,4,54,56,6,3,2,1)
```

```python
add_nums(1,2,3,3,4,54,56,6,3,2,1,'a')
```

```python
def fib_series(n):
    if n <=1:
        return 1
    
    else:
        return fib_series(n-1) + fib_series(n-2)
```

```python
fib_series(10)/fib_series(9)
```
