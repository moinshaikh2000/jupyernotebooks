```python
train,test = image_dataset_from_directory(directory=r"C:\Users\shaik\OneDrive\Desktop\Pyton Class\Data set\AI Image",batch_size=32,
                                image_size=(64, 64),validation_split=0.2,subset="both",seed=1,
                                         label_mode='binary')
```

```python
for i,j in test.take(1):
    print(j)
```

```python
model = Sequential()
model.add(Conv2D(input_shape=(64,64,3),kernel_size=(3,3),filters=64,activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Conv2D(kernel_size=(3,3),filters=128,activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(train,epochs=10)
```
